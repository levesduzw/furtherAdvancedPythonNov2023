```python
# Group the data into separate data frames for each airline (pick a few)
df_AA = df[df['carrier'] == 'AA']
df_AA
df_AS = df[df['carrier'] == 'AS']
df_AS
df_US = df[df['carrier'] == 'US']
df_US
df_DL = df[df['carrier'] == 'DL']
df_DL
df_UA = df[df['carrier'] == 'UA']
df_UA
```

```python
# Show the minimum and mean departure delay for your chosen airlines (or all)
airline_df_list = [df_AA, df_AS, df_US, df_DL, df_UA]

airline_stats = pd.DataFrame(index=['AA', 'AS', 'US', 'DL', 'UA'])
airline_stats
```

```python
data = [ airline_df['dep_delay'].clip(lower=0).min() for airline_df in airline_df_list ]

airline_stats['Minimum departure delay'] = data
airline_stats
```

```python
df_AA['dep_time'].mean()
```

```python
data = [ airline_df['dep_delay'].mean() for airline_df in airline_df_list ]

airline_stats['Mean departure delay'] = data
airline_stats
```

```python
# number of flight records

data = [ airline_df['flight'].count() for airline_df in airline_df_list ]

airline_stats['Number of flight records'] = data
airline_stats
```

```python
# How many unique aircraft are represented (tail number)

data = [ airline_df['tailnum'].nunique() for airline_df in airline_df_list ]

airline_stats['Number of unique aircraft'] = data
airline_stats
```

```python
(df_AA['dep_delay'] + df_AA['arr_delay']).describe()
```

```python
# The greatest recorded delay (show the related data) 

data = [ (airline_df['dep_delay']+airline_df['arr_delay']).max() for airline_df in airline_df_list ]

airline_stats['Greatest recorded delay'] = data
airline_stats
```

```python
# The mean values for just the first 50 records in the dataset

data = [ (airline_df['dep_delay']+airline_df['arr_delay']).head(50).mean() for airline_df in airline_df_list ]

airline_stats['Mean of first 50 delay records'] = data
airline_stats
```

```python
# then for all delays of 15 or more units

data = []

for airline_df in airline_df_list:
    delay_series = airline_df['dep_delay'] + airline_df['arr_delay']
    delay_series = delay_series[delay_series > 15]
    data.append(delay_series.mean())
    #airlife_sorted_df = df.sort_values(by='average',ascending=False, inplace=True)

    
airline_stats['Mean of 15m+ delays'] = data
airline_stats
```

```python
 #Which airline arrives ahead of time most often

data = []

for airline_df in airline_df_list:
    arr_delay_series = airline_df['arr_delay']
    less_than_0_count = arr_delay_series.lt(0).sum()

    data.append(less_than_0_count)

#airline_stats.insert(0, '% Arrived ahead of time 2', data)
airline_stats['Arrived ahead of time'] = data
airline_stats
```

```python
 #Which airline arrives ahead of time most often

data = []

for airline_df in airline_df_list:
    arr_delay_series = airline_df['arr_delay']
    less_than_0_count = arr_delay_series.lt(0).sum()
    
    airline_flight_count = airline_df['flight'].count()
    
    data.append(less_than_0_count / airline_flight_count * 100)

#airline_stats.insert(0, '% Arrived ahead of time 2', data)
airline_stats['% Arrived ahead of time'] = data
airline_stats
```
